```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
from sklearn.cluster import KMeans
```

```python
data = pd.read_csv('/content/3.12.+Example (1).csv')
```

```python
data
```

```python
plt.scatter(data['Satisfaction'],data['Loyalty'])
plt.xlabel('Satisfaction')
plt.ylabel('Loyalty')
```

```python
x=data.copy()
```

```python
kmeans=KMeans(2)
kmeans.fit(x)
```

```python
clusters = x.copy()
clusters['cluster_pred'] = kmeans.fit_predict(x)
```

```python
plt.scatter(clusters['Satisfaction'],clusters['Loyalty'],c=clusters['cluster_pred'],cmap='rainbow')
plt.xlabel('Satisfaction')
plt.ylabel('Loyalty')
```

```python
from sklearn import preprocessing
X_scaled=preprocessing.scale(x)
X_scaled
```

```python
wcss=[]
for i in range(1,10):
  kmeans=KMeans(i,n_init='auto')
  kmeans.fit(X_scaled)
  wcss.append(kmeans.inertia_)
wcss
```

```python
plt.plot(range(1,10),wcss)
plt.xlabel('no. of clusters')
plt.ylabel('wcss')
```

```python
kmeans_new = KMeans(4)
kmeans_new.fit(X_scaled)
clusters_new = x.copy()
clusters_new['clusters_pred']= kmeans_new.fit_predict(X_scaled)
clusters_new
```

```python
plt.scatter(clusters_new['Satisfaction'],clusters_new['Loyalty'],c=clusters_new['clusters_pred'],cmap='rainbow')
plt.xlabel('Satisfaction')
plt.ylabel('Loyalty')
```
